```python
tools = []
for i in new_data['工具']:
    if i == 0:
        tools.append('others')
    else:
        if i[-1] == 'd':
    #         i = 'Android'
            tools.append('Android')
        elif i[-1] == 'm':
    #         i = 'weibo.com'
            tools.append('weibo.com')
        elif i[-1] == '话':
    #         i = '超话'
            tools.append('超话')

        else:
    #         i = 'others'
            tools.append('others')
```

```python
at_num = []
for i in new_data['@用户']:
    if i == 0:
        at_num.append(0)
    else:
        at_num.append(int(i.count(',')+1))
```
